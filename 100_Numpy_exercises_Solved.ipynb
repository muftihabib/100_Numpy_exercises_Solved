#### 75. How to compute averages using a sliding window over an array? (★★★)

````python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
    Z = np.arange(20)
print(moving_average(Z, n=3))
````

#### 76. Consider a one-dimensional array Z, build a two-dimensional array whose first row is (Z[0],Z[1],Z[2]) and each subsequent row is  shifted by 1 (last row should be (Z[-3],Z[-2],Z[-1]) (★★★)

````python
from numpy.lib import stride_tricks
def rolling(a, window):
    shape = (a.size - window + 1, window)
    strides = (a.strides[0], a.strides[0])
    return stride_tricks.as_strided(a, shape=shape, strides=strides)
Z = rolling(np.arange(10), 3)
print(Z)
````

#### 78. Consider 2 sets of points P0,P1 describing lines (2d) and a point p, how to compute distance from p to each line i (P0[i],P1[i])? (★★★)

````python
def distance(P0, P1, p):
    T = P1 - P0
    L = (T**2).sum(axis=1)
    U = -((P0[:,0]-p[...,0])*T[:,0] + (P0[:,1]-p[...,1])*T[:,1]) / L
    U = U.reshape(len(U),1)
    D = P0 + U*T - p
    return np.sqrt((D**2).sum(axis=1))
P0 = np.random.uniform(-10,10,(10,2))
P1 = np.random.uniform(-10,10,(10,2))
p  = np.random.uniform(-10,10,( 1,2))
print(distance(P0, P1, p))
````

#### 85. Create a 2D array subclass such that Z[i,j] == Z[j,i] (★★★)

````python
class Symmetric(np.ndarray):
    def __setitem__(self, index, value):
        i, j = index
        super(Symmetric, self).__setitem__((i, j), value)
        super(Symmetric, self).__setitem__((j, i), value)

def create_symmetric(Z):
    sym_matrix = Z + Z.T - np.diag(Z.diagonal())
    return np.asarray(sym_matrix).view(Symmetric)
Z = np.random.randint(0, 10, (5, 5))
S = create_symmetric(Z)

print("Initial Symmetric Matrix:")
print(S)
S[2, 3] = 999

print("\nMatrix after setting S[2, 3] = 999:")
print(S) 
````

#### 88. How to implement the Game of Life using numpy arrays? (★★★)

````python
def iterate(Z):
    N = (Z[0:-2, 0:-2] + Z[0:-2, 1:-1] + Z[0:-2, 2:] +
         Z[1:-1, 0:-2]                 + Z[1:-1, 2:] +
         Z[2:  , 0:-2] + Z[2:  , 1:-1] + Z[2:  , 2:])
    
    birth = (N == 3) & (Z[1:-1, 1:-1] == 0)
    survive = ((N == 2) | (N == 3)) & (Z[1:-1, 1:-1] == 1)
    
    Z[...] = 0
    Z[1:-1, 1:-1][birth | survive] = 1
    return Z

Z = np.random.randint(0, 2, (10, 10))

for _ in range(5): 
    Z = iterate(Z)

print(Z)
````

````python
def cartesian_numpy(*arrays):
    grid = np.meshgrid(*arrays, indexing='ij')
    return np.stack(grid, axis=-1).reshape(-1, len(arrays))
v1 = np.array([1, 2])
v2 = np.array([3, 4])
v3 = np.array([5, 6])
result = cartesian_numpy(v1, v2, v3)
print(result)
````

#### 98. Considering a path described by two vectors (X,Y), how to sample it using equidistant samples (★★★)?

````python
phi = np.arange(0, 10 * np.pi, 0.5)
X = phi * np.cos(phi)
Y = phi * np.sin(phi)
dx = np.diff(X)
dy = np.diff(Y)
distances = np.sqrt(dx**2 + dy**2)
cumulative_distance = np.insert(np.cumsum(distances), 0, 0)
num_samples = 200
equidistant_steps = np.linspace(0, cumulative_distance[-1], num_samples)
X_sampled = np.interp(equidistant_steps, cumulative_distance, X)
Y_sampled = np.interp(equidistant_steps, cumulative_distance, Y)
print(f"Original number of points: {len(X)}")
print(f"New number of equidistant points: {len(X_sampled)}")
````

#### 99. Given an integer n and a 2D array X, select from X the rows which can be interpreted as draws from a multinomial distribution with n degrees, i.e., the rows which only contain integers and which sum to n. (★★★)

````python
X = np.array([[1.0, 0.0, 3.0, 8.0],   
              [2.0, 0.0, 1.0, 1.0],   
              [1.5, 2.5, 0.0, 0.0],     
              [5.0, -1.0, 0.0, 0.0]]) 
n = 4
is_integer = np.all(np.mod(X, 1) == 0, axis=1)
sums_to_n = (np.sum(X, axis=1) == n)
is_non_negative = np.all(X >= 0, axis=1)
valid_rows = is_integer & sums_to_n & is_non_negative
result = X[valid_rows]
print("Original Matrix:\n", X)
print(f"\nRows that sum to {n} and contain only non-negative integers:\n", result)
````

#### 100. Compute bootstrapped 95% confidence intervals for the mean of a 1D array X (i.e., resample the elements of an array with replacement N times, compute the mean of each sample, and then compute percentiles over the means). (★★★)

````python
X = np.random.randn(100)
N = 1000
bootstrap_samples = np.random.choice(X, size=(N, len(X)), replace=True)
bootstrap_means = np.mean(bootstrap_samples, axis=1)
confidence_interval = np.percentile(bootstrap_means, [2.5, 97.5])
print("Original Array Mean:", np.mean(X))
print("Bootstrapped 95% Confidence Interval:", confidence_interval)
````
